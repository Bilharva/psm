este notebook realiza a leitura, análise e visualização dos dados climáticos de um único dia contidos no arquivo `clima.csv`.

## Parte 1: Leitura e Organização dos Dados

```python
# importação da biblioteca de gráficos
import matplotlib.pyplot as plt

# criação das quatro listas que vão armazenar os dados
hora = []
temperatura = []
umidade = []
precipitacao = []

# abertura do arquivo no modo leitura
arquivo = open('clima.csv', 'r')

# leitura linha por linha com laço for
for linha in arquivo:

    # remove espaços em branco e quebras de linha
    linha = linha.strip()

    # pula a linha de cabeçalho
    if linha.startswith('hora'):
        continue

    # separa os valores usando a vírgula como delimitador
    valores = linha.split(',')

    # converte cada valor para o tipo correto e armazena nas listas
    hora.append(int(valores[0]))
    temperatura.append(float(valores[1]))
    umidade.append(float(valores[2]))
    precipitacao.append(float(valores[3]))

arquivo.close()

print('dados lidos com sucesso!')
print(f'total de registros: {len(hora)}')
```

## Parte 2: Análise dos Dados

```python
# temperatura máxima
temp_max = max(temperatura)
hora_temp_max = hora[temperatura.index(temp_max)]
print(f'temperatura máxima: {temp_max}°C às {hora_temp_max}h')

#temperatura mínima
temp_min = min(temperatura)
hora_temp_min = hora[temperatura.index(temp_min)]
print(f'temperatura mínima: {temp_min}°C às {hora_temp_min}h')

# maior umidade
umidade_max = max(umidade)
hora_umidade_max = hora[umidade.index(umidade_max)]
print(f'maior umidade: {umidade_max}% às {hora_umidade_max}h')

# total acumulado de precipitação
total_chuva = sum(precipitacao)
print(f'total de precipitação no dia: {total_chuva:.1f} mm')

#  horario de maior precipitação
precip_max = max(precipitacao)
hora_precip_max = hora[precipitacao.index(precip_max)]
print(f'maior precipitação: {precip_max} mm às {hora_precip_max}h')
```

## Parte 3: Construção dos Gráficos

```python
#  gráfico 1: hora x temperatura (linha vermelha)

plt.plot(hora, temperatura, color='red', linewidth=2)

# configuração do grid: tracejado, cinza, opacidade 0.5
plt.grid(linestyle='--', color='gray', alpha=0.5)

# títulos e rótulos dos eixos
plt.title('Temperatura ao longo do dia')
plt.xlabel('Hora do dia')
plt.ylabel('Temperatura (°C)')

plt.show()
```

```python
# gráfico 2: Hora x Umidade (linha azul)

plt.plot(hora, umidade, color='blue', linewidth=2)

# configuração do grid: pontilhado, verde claro, opacidade 0.5
plt.grid(linestyle=':', color='lightgreen', alpha=0.5)

# títulos e rótulos dos eixos
plt.title('Umidade relativa ao longo do dia')
plt.xlabel('Hora do dia')
plt.ylabel('Umidade (%)')

plt.show()
```

```python
# --- gráfico 3: hora x precipitação (barras roxas)

plt.bar(hora, precipitacao, color='purple', alpha=0.7)

# configuração do grid: tracejado, preto, opacidade 0.5
plt.grid(linestyle='--', color='black', alpha=0.5)

# títulos e rótulos dos eixos
plt.title('Precipitação ao longo do dia')
plt.xlabel('Hora do dia')
plt.ylabel('Precipitação (mm)')

plt.show()
```
